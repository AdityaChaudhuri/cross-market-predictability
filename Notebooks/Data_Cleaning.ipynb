```python
import pandas as pd
import yfinance as yf
```

```python
# Download historical stock data for the specified tickers and date range

tickers = ["AAPL", "JPM", "XOM", "MC.PA", "ASML.AS", "BHP.AX", "7203.T", "RELIANCE.NS", "VALE3.SA", 'MSCI']

stock_data = yf.download( tickers , start = "2020-01-01", end = "2025-12-31", group_by = 'ticker')

stock_df = pd.DataFrame(stock_data)
print(stock_df.head())
```

```python
# Download historical data for the specified reigons

regions = ["^GSPC", "^IXIC", "^RUT", "^GDAXI", "^STOXX50E", "^FCHI", "^N225", "^HSI", "^AXJO", "^KS11", "^BSESN", "^BVSP", "EEM", "^FTSE"]

regions_data = yf.download(regions, start="2020-01-01", end="2025-12-31", group_by="ticker")

regions_df = pd.DataFrame(regions_data)
print(regions_df.head())
```

```python
#basic info about the datasets

print("shape of stock_df: ", stock_df.shape)
print("shape of regions_df: ", regions_df.shape)

print("columns of stock_df: ", stock_df.columns)
print("columns of regions_df: ", regions_df.columns)
```

```python
#checking for missing values in both datasets

print ("stock dataset missing values: ", stock_df.isnull().sum())

print ("regions dataset missing values: ", regions_df.isnull().sum())
```

```python
#cleaning missing data by forward filling

stock_df_filled = stock_df.ffill()
regions_df_filled = regions_df.ffill()

print ("stock dataset missing values: ", stock_df_filled.isnull().sum())

print ("regions dataset missing values: ", regions_df_filled.isnull().sum())
```

```python
#cleaning last remaining missing values by backward filling

stock_df_filled = stock_df_filled.bfill()
regions_df_filled = regions_df_filled.bfill()

print ("stock dataset missing values: ", stock_df_filled.isnull().sum())
print ("regions dataset missing values: ", regions_df_filled.isnull().sum())
```

```python
#converting all currencies to USD using exchange rates from Yahoo Finance

fx_tickers = ["EURUSD=X", "GBPUSD=X", "AUDUSD=X", "JPYUSD=X", "INRUSD=X", "BRLUSD=X", "HKDUSD=X", "KRWUSD=X"]

fx_data = yf.download(fx_tickers, start="2020-01-01", end="2025-12-31", group_by="ticker")
```

```python
fx_df = pd.DataFrame(fx_data)
print(fx_df.head())
print("shape of fx_df: ", fx_df.shape)
```

```python
print("missing values in fx_df: ", fx_df.isnull().sum())
```

```python
# extracting only the 'Close' prices for the exchange rates to use for currency conversion
fx_close = fx_df.xs("Close", level=1, axis=1)

# forward fill and backward fill any remaining missing values in the exchange rates
fx_close_filled = fx_close.ffill().bfill()

# Define mapping of tickers to their FX pair
fx_mapping = {
    'MC.PA': 'EURUSD=X',
    'ASML.AS': 'EURUSD=X',
    'BHP.AX': 'AUDUSD=X',
    '7203.T': 'JPYUSD=X',
    'RELIANCE.NS': 'INRUSD=X',
    'VALE3.SA': 'BRLUSD=X'
}

region_fx_mapping = {
    '^AXJO': 'AUDUSD=X',
    '^N225': 'JPYUSD=X',
    '^BSESN': 'INRUSD=X',
    '^HSI': 'HKDUSD=X',
    '^KS11': 'KRWUSD=X',
    '^FTSE': 'GBPUSD=X',
    '^STOXX50E': 'EURUSD=X',
    '^GDAXI': 'EURUSD=X',
    '^FCHI': 'EURUSD=X',
    '^BVSP': 'BRLUSD=X'
}

# Convert stock prices to USD

stock_df_converted = stock_df_filled.copy()
for ticker, fx_pair in fx_mapping.items():
    for price_type in ['Open', 'High', 'Low', 'Close']:
        stock_df_converted[(ticker, price_type)] = (
            stock_df_filled[(ticker, price_type)] * fx_close_filled[fx_pair]
        )

# Convert region index values to USD
regions_df_converted = regions_df_filled.copy()
for ticker, fx_pair in region_fx_mapping.items():
    for price_type in ['Open', 'High', 'Low', 'Close']:
        regions_df_converted[(ticker, price_type)] = (
            regions_df_filled[(ticker, price_type)] * fx_close_filled[fx_pair]
        )


print("USD normalisation complete")
```

```python
from pathlib import Path

data_dir = Path("../Data")
data_dir.mkdir(parents=True, exist_ok=True)

stock_df_converted.to_csv(data_dir / "stock_data_converted.csv")
regions_df_converted.to_csv(data_dir / "regions_data_converted.csv")

print(f"Datasets saved to {data_dir.resolve()}")
```
